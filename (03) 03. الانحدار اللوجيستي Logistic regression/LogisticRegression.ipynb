```python
import pandas as pd

df = pd.read_csv('income.csv')

x = df[['Age', 'Education']]
y = df['VIP']
```

```python
dfVIP = df[df['VIP'] == 1]

dfNotVIP = df[df['VIP'] == 0]

import matplotlib.pyplot as plt

plt.scatter(dfVIP['Education'], dfVIP['Age'], color='red', marker='x')
plt.scatter(dfNotVIP['Education'], dfNotVIP['Age'], color='blue', marker='x')
plt.legend(['VIP', 'Not VIP'])
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x, y)

b0 = model.intercept_
b1 = model.coef_[0]
b2 = model.coef_[1]

print('Intercept (b0):', round(b0,2))
print('Coefficient (b1):', round(b1,2))
print('Coefficient (b2):', round(b2,2))
```

```python
import math

def sigmoid(z):
    return 1 / (1 + pow(math.e, -z))

Age = 60
Education = 12
Boundary = 0.5

y_lr = b0 + b1 * Age + b2 * Education

sig_y = sigmoid(y_lr)

if sig_y > Boundary:
    print('VIP')
else:
    print('Not VIP')
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(x, y)
```

```python
Age = 60
Education = 12

example = [[Age, Education]]

income_info = pd.DataFrame(example, columns=['Age', 'Education'])

pred = model.predict(income_info)

if pred == 1:
    print('VIP')
else:
    print('Not VIP')
```

```python
p_proba = model.predict_proba(income_info)
print(p_proba)
```
