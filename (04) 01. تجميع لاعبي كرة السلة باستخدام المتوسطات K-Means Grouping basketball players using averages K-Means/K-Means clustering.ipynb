```python
import matplotlib.pyplot as plt
import os

x = [[5,3],
[10,15],
[15,12],
[24,10],
[30,45],
[85,70],
[71,80],
[60,78],
[55,52],
[80,91]]

x1 = [i[0] for i in x]
x2 = [i[1] for i in x]
plt.scatter(x1,x2)
plt.show()
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=2)
model.fit(x)

print('Center: ', model.cluster_centers_)
print('Labels: ', model.labels_)
```

```python
x1 = [i[0] for i in x]
x2 = [i[1] for i in x]
colors = model.labels_
plt.scatter(x1,x2, c=colors)

cx1 = model.cluster_centers_[:,0]
cx2 = model.cluster_centers_[:,1]
plt.scatter(cx1,cx2, color='green')
print(type(model.cluster_centers_))
```

```python
from sklearn.metrics import silhouette_score
silhouette_avg = silhouette_score(x, model.labels_)
print(round(silhouette_avg, 2))
```

```python
model = KMeans(n_clusters=3)
model.fit(x)

plt.scatter([i[0] for i in x], [i[1] for i in x], c=model.labels_, cmap='rainbow')
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], color='black')
```

```python
silhouette_avg = silhouette_score(x, model.labels_)
print(round(silhouette_avg,2))
```

```python
# Set OMP_NUM_THREADS to avoid memory leak on Windows with MKL
os.environ['OMP_NUM_THREADS'] = '1'

range_clusters = [2, 3, 4, 5, 6, 7] # number 2 has the best value because it is nearer to 1
for n_clusters in range_clusters:
    
    model = KMeans(n_clusters=n_clusters, n_init=10)  # Explicitly setting n_init to suppress FutureWarning
    model.fit(x)
    silhouette_avg = silhouette_score(x, model.labels_)
    print('For n_clusters =', n_clusters,'The average silhouette_score is :', round(silhouette_avg,2))
```
