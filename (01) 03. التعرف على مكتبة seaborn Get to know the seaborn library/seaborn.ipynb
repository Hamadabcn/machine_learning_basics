```python
import pandas as pd

df = pd.read_csv('titanic.csv')

df.head(10)
```

```python
import seaborn as sns

sns.countplot(x='Sex', data=df)
```

```python
sns.countplot(x='Sex', data=df, hue='Survived')
```

```python
sns.displot(x='Age', data=df, bins=10, color='blue')
```

```python
sns.displot(x='Age', data=df, bins=10, color='blue', hue=df['Survived'], kde=True)
```

```python
sns.lineplot(x=df['Age'], y=df['Weight'])
```

```python
sns.lineplot(x=df['Age'], y=df['Weight'], hue=df['Sex'])
```

```python
sns.scatterplot(x=df['Age'], y=df['Weight'])
```

```python
sns.scatterplot(x=df['Age'], y=df['Weight'], hue=df['Sex'])
```

```python
sns.boxplot(data=df['Age'])
```

```python
sns.boxplot(x=df['Sex'], y=df['Age'])
```

```python
sns.barplot(x='Pclass', y='Age', data=df)
```
