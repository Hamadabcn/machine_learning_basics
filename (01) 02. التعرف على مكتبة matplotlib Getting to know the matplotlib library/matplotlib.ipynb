```python
import pandas as pd

df = pd.read_csv('failures.csv')

df
#df.info()
```

```python
import matplotlib.pyplot as plt

xpoints = df['Age']
ypoints = df['Failures']

plt.plot(xpoints, ypoints, marker='o', color='red')

plt.xlabel('Age')
plt.ylabel('Failures')

plt.show()
```

```python
x = df['Machine']
y = df['Failures']

plt.xlabel('Machine')
plt.ylabel('Faiulures')

plt.bar(x,y)
plt.show()
```

```python
x = df['Machine']
y = df['Failures']

plt.pie(y, labels = x, autopct='%1.1f%%')
plt.show()
```

```python
xpoints = df['Age']
y1points = df['Failures']
y2points = df['DaysOff']

plt.subplot(1,2,1)
plt.scatter(xpoints, y1points)

plt.xlabel('Age')
plt.ylabel('Failures')

plt.subplot(1,2,2)
plt.scatter(xpoints, y2points)

plt.xlabel('Age')
plt.ylabel('DaysOff')

plt.show()
```

```python
df = pd.read_csv('stdMarks.csv')

dfcorrel = df[['Math', 'Phy', 'Lang']]

correlations = dfcorrel.corr(method='pearson')
print(correlations)
```

```python
fig = plt.figure()

ax = fig.add_subplot(1,1,1)

cax = ax.matshow(correlations, vmin=-1, vmax=1)

fig.colorbar(cax)

ticks = [0,1,2]

ax.set_xticks(ticks)
ax.set_yticks(ticks)

ax.set_xticklabels(['Math', 'Phy', 'Lang'])
ax.set_yticklabels(['Math', 'Phy', 'Lang'])

plt.show()
```
