```python
import pandas as pd

df = pd.read_csv('income.csv')

df.info()
```

```python
df.describe()
```

```python
df = df[['Age', 'Education', 'Gender', 'Income']]

from pandas.plotting import scatter_matrix

scatter_matrix(df, figsize=(8,8))
```

```python
x = df[['Age', 'Education']]
y = df['Income']

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x_train, y_train)

print('Intercept:', model.intercept_)
print('Coefficients', model.coef_)
```

```python
y_pred = model.predict(x_test)

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from math import sqrt

print('MSE :', round(mean_absolute_error(y_pred, y_test), 2))
print('RMSE :', round(mean_squared_error(y_pred, y_test), 2))
print('MAE :', round(mean_absolute_error(y_pred, y_test), 2))
```
